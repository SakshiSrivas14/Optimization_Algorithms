#### Use your code from practical session 1 and create a function to perform GD for single variable linear regression.
#### The function should take the following input parameters:
##### Input data (X), Target Labels (y), Learning Rate (alpha), Maximum number of iterations
#### The function should return the following:
##### All Theta_0 and Theta_1 values till converge, and equivellant Losses, all hypothesis output, and optimum values of Theta_0 and Theta_1
#### Note: you can choose any suitable stop criteria.

```python
def GD(x, y,alpha,max_iter) :
    i = 0
    theta_0=0
    theta_1=0
    costList=[]
    theta_0_lst=[]
    theta_1_lst=[]
    m=50
    while i<max_iter:
        h = x*theta_1 + theta_0
        #calculate cost function
        cost = (1/(2*m))  * np.sum((h - y) ** 2)
        

        #calculate theta0
        gradient_0 = np.sum(h - y) / m
        theta_0 = theta_0 - alpha * gradient_0
        
        
        #calculate theta1
        gradient_1 = np.sum((h - y)*(x))   / m
        theta_1 = theta_1 - alpha * gradient_1 
        print("Iteration %d | Cost: %f" % (i,cost))
        #stop condition:
        if (gradient_0**2 + gradient_1**2)**0.5 <0.05:
            break
        i += 1
        costList.append(cost)
        theta_0_lst.append(theta_0)
        theta_1_lst.append(theta_1)    
    return theta_0_lst , theta_1_lst , h ,costList
```

#### Update the implementation of the Batch GD to be Mini-Batch and rpeat the above plots and evaluations
#### Change batch size and see the effect

```python
def MGD(x, y, alpha, max_itter,batch_size) :
    i = 0
    theta_0=0
    theta_1=0
    costList=[]
    theta_0_lst=[]
    theta_1_lst=[]

    m=len(y)
    n_batches = int(np.ceil(m/batch_size))
    
    for i in range(max_itter):
        for itr in range(0, m,batch_size):
            start_index = itr
            end_index = (itr + batch_size)
          
            x_= x[start_index:end_index]
            y_= y[start_index:end_index]
            
            size=len(x_)
            h = x_*theta_1 + theta_0
            #calculate cost function
            cost = (1/(2*size))  * np.sum((h - y_) ** 2)
            
            #print("Iteration %d | Cost: %f" % (i,cost))

            #calculate theta0
            gradient_0 = np.sum(h - y_) / size
            theta_0 = theta_0 - alpha * gradient_0

            #calculate theta1
            gradient_1 = np.sum((h - y_)*(x_))   / size
            theta_1 = theta_1 - alpha * gradient_1 
            print("Iteration %d | Cost: %f | theta0:%f | theta1:%f" % (i+1,cost,theta_0,theta_1))
            costList.append(cost)
            theta_0_lst.append(theta_0)
            theta_1_lst.append(theta_1)
        #stop condition:Norm2
        if (gradient_0**2 + gradient_1**2)**0.5 <0.05:
            break
        i += 1    
    return theta_0_lst , theta_1_lst , h ,costList
```

## Stochastic GD

#### Update the implementation of the Batch GD to beStochastic and rpeat the above plots and evaluations
#### You can try different learning rate and see its effect.
#### Compare to Batch and Mini-Batch GD

```python
def SGD(x, y, alpha,max_iter) :
    i = 0
    theta_0=0 
    theta_1=0
    costList=[]
    theta_0_lst=[]
    theta_1_lst=[]
    m=len(y)
    
    for j in range(max_iter):
        for i in range(m):
            h = x[i]*theta_1 + theta_0
            #calculate cost function
            cost = (1/(2))  * ((h - y[i]) )** 2
            #print("Iteration %d | Cost: %f" % (i,cost))

            #calculate theta0
            gradient_0 = (h - y[i]) 
            theta_0 = theta_0 - alpha * gradient_0

            #calculate theta1
            gradient_1 = (h - y[i])*(x[i])  
            theta_1 = theta_1 - alpha * gradient_1 
            print("Iteration %d | Cost: %f | theta0:%f | theta1:%f" % (i+1,cost,theta_0,theta_1))

            costList.append(cost)
            theta_0_lst.append(theta_0)
            theta_1_lst.append(theta_1)
    
    #stop condition:
        if (gradient_0**2 + gradient_1**2)**0.5 <0.05:
            break
        
      
    return theta_0_lst , theta_1_lst , h  , costList 
```
