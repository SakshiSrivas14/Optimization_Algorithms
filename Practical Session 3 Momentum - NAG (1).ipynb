#### Use your final version of the implemented Batch GD algorithm for single variable to find the best line filt of the given data.
#### Coose the suitable number of iterations, learning rate, and stop criteria.
#### Calculate r2 score. Shouldn't below 0.9
#### Plot the required curves (loss-epochs, loss-theta0, loss-theta1, all fitted lines per epoch (single graph) , best fit line)

```python
def GD(x, y,alpha, max_iter) :
    i = 0
    theta_0=0
    theta_1=0
    costList=[]
    theta_0_lst=[]
    theta_1_lst=[]
    m=50
    while i<max_iter:
        h = x*theta_1 + theta_0
        #calculate cost function
        cost = (1/(2*m))  * np.sum((h - y) ** 2)
    
        #calculate theta0
        gradient_0 = np.sum(h - y) / m
        theta_0 = theta_0 - alpha * gradient_0
        
        
        #calculate theta1
        gradient_1 = np.sum((h - y)*(x))   / m
        theta_1 = theta_1 - alpha * gradient_1 
        print("Iteration %d | Cost: %f" % (i,cost))
        #stop condition:
        if (gradient_0**2 + gradient_1**2)**0.5 <0.05:
            break
        i += 1
        costList.append(cost)
        theta_0_lst.append(theta_0)
        theta_1_lst.append(theta_1)
       
    return theta_0_lst , theta_1_lst , h , costList
```

```python
theta_0_lst , theta_1_lst , h , gradient_0,gradient_1,costList=GD(x, y, 0.001,1000)
```

```python
y_pred=theta_0_lst[-1]+ theta_1_lst[-1]*x
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y, y_pred)
```

```python
plt.plot(costList)
```

```python
plt.plot(theta_0_lst,costList)
plt.show()
```

```python
plt.plot(theta_1_lst,costList)
plt.show()
```

### Update your Batch GD for one variable  implementation to be Momentum-Based GD and check your results

#### γ takes values between 0 and 1.


```python
def momentum_GD(x, y,alpha,max_iter,gamma) :
    i = 0
    theta_0=0
    theta_1=0
    vt0=0
    vt1=0
    
    costList=[]
    theta_0_lst=[]
    theta_1_lst=[]
    m=50
    while i<max_iter:
        h = x*theta_1 + theta_0
        #calculate cost function
        cost = (1/(2*m))  * np.sum((h - y) ** 2)
    
        #calculate theta0
        gradient_0 = np.sum(h - y) / m
        theta_0 = theta_0 - alpha * gradient_0
        vt0=gamma * vt0 + alpha * gradient_0 
        theta_0= theta_0 - vt0
        
        #calculate theta1
        gradient_1 = np.sum((h - y)*(x))   / m
        theta_1 = theta_1 - alpha * gradient_1 
        vt1=gamma * vt1 + alpha * gradient_1 
        theta_1= theta_1 - vt1
        print("Iteration %d | Cost: %f" % (i,cost))
        #stop condition:
        if (gradient_0**2 + gradient_1**2)**0.5 <0.05:
            break
        i += 1
        costList.append(cost)
        theta_0_lst.append(theta_0)
        theta_1_lst.append(theta_1)
       
    return theta_0_lst , theta_1_lst , h , costList
```

```python
theta_0_lst , theta_1_lst , h , costList= momentum_GD(x, y, 0.0005,1000,0.9)
costList[-1]
```

### Update your Batch GD for one variable  implementation to be NAG and check your results

```python
def NGD(x, y, alpha,max_iter,gamma) :
    i = 0
    theta_0=0 
    theta_1=0
    vt0=0
    vt1=0
    theta_0_temp=0
    theta_1_temp=0
    costList=[]
    theta_0_lst=[]
    theta_1_lst=[]
    m=50
    while i<max_iter:
    
        #calculate theta0
        theta_0_temp= theta_0 - gamma*vt0
        theta_1_temp = theta_1 -gamma*vt1
        h=theta_0_temp + theta_1_temp*x
        vt0=gamma * vt0 + alpha*np.sum(h-y) / m
        
        vt1=gamma * vt1 + alpha* np.sum((h - y)*(x))   / m
        theta_0 = theta_0 - vt0
        theta_1 =theta_1 - vt1
        
        cost = (1/(2*m))  * np.sum((h - y) ** 2)
        
        #stop condition:
        if (gradient_0**2 + gradient_1**2)**0.5 <0.05:
            break
        i += 1
        costList.append(cost)
        theta_0_lst.append(theta_0)
        theta_1_lst.append(theta_1)
       
    return theta_0_lst , theta_1_lst , h , costList
```
